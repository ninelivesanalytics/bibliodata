```python
from pathlib import Path

# 1) Detectar raíz del repo (bibliodata/)
cwd = Path.cwd()

# Si estás en .../bibliodata/notebooks o .../bibliodata/notebooks/algo
if "notebooks" in [p.name for p in cwd.parents] or cwd.name == "notebooks":
    # sube hasta /notebooks y luego a su padre (raíz del repo)
    notebooks_dir = cwd if cwd.name == "notebooks" else next(p for p in cwd.parents if p.name == "notebooks")
    ROOT = notebooks_dir.parent
else:
    # fallback: asume que ya estás en la raíz del repo
    ROOT = cwd

print("Root detectada:", ROOT)

# 2) Crear estructura
DIRS = [
    "data/raw",
    "data/interim",
    "data/processed",
    "data/samples",
    "docs",
    "sql/schema",
    "sql/seeds",
    "sql/queries",
    "graph/cypher",
    "src/config",
    "src/ingest",
    "src/nlp",
    "src/storage",
    "src/api/routes",
    "src/pipelines",
    "tests",
]

for d in DIRS:
    (ROOT / d).mkdir(parents=True, exist_ok=True)

# 3) .gitkeep para carpetas que normalmente irán ignoradas
GITKEEP_DIRS = [
    "data/raw",
    "data/interim",
    "data/processed",
]

for d in GITKEEP_DIRS:
    (ROOT / d / ".gitkeep").touch(exist_ok=True)

print("Estructura creada.")
```
